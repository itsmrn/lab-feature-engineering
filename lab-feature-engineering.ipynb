```python
# Checking to see if nan values have a big impact, for the WEALTH1 col the proportion of nan values is too high, 
# we will delete this col to not bring in bias into our data, the rest of nans we will replace by mean values. 

for i in selected:
    a = selected[i].value_counts(dropna=False)/len(selected)
    print("---------------")
    print (i)
    print (a)
```
